```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("../data/twitter.csv")
df.head()
```

```python
print(df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={"sentence": "text", "sentiment": "label"})
df.head()
```

```python
sns.countplot(x="label", data=df)
plt.title("Sentiment Distribution")
plt.savefig("../images/sentiment_distribution.png")
plt.show()
```

```python
df["text_length"] = df["text"].apply(len)

sns.histplot(df["text_length"], bins=50)
plt.title("Text Length Distribution")
plt.savefig("../images/text_length.png")
plt.show()
```

```python
df["word_count"] = df["text"].apply(lambda x: len(str(x).split()))

sns.histplot(df["word_count"], bins=50)
plt.title("Word Count Distribution")
plt.savefig("../images/word_count.png")
plt.show()
```

```python
from collections import Counter

all_words = " ".join(df["text"]).split()
common_words = Counter(all_words).most_common(20)

words = [w[0] for w in common_words]
counts = [w[1] for w in common_words]

plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 20 Words")
plt.savefig("../images/top_words.png")
plt.show()
```

```python
positive = df[df["label"] == 1]
negative = df[df["label"] == 0]

pos_words = " ".join(positive["text"]).split()
neg_words = " ".join(negative["text"]).split()

print("Top Positive Words:", Counter(pos_words).most_common(10))
print("Top Negative Words:", Counter(neg_words).most_common(10))
```
